#### Exponents:

An exponent tells you how many times to multiply a base number by itself. 
It is written as a small superscript number next to the base

```python
#Basic Example: \(5^{3}\) (five to the third power) means \(5 \times 5 \times 5 = 125\)
Five = 5
print(f" Five at the power of 3 is: {(Five ** 2)}")

# Zero Rule: Any non-zero number raised to the power of \(0\) equals \(1\) (e.g., \(18^0 = 1\)).
print(f" Five at the power of 2 is: {(Five ** 0)}")

# Negative Exponents: A negative exponent indicates a reciprocal. For example, \(3^{-2} = \frac{1}{3^2} = \frac{1}{9}\).
print(f" Five at the power of 3 is: {(Five ** -2)}")
```

#### Roots:

A root is the opposite of an exponent. 

Finding the square root of \(64\) is the same as asking: "What number multiplied by itself equals \(64\)?"

```python
# Square Roots: Written as \(\sqrt{x}\), this implies an invisible index of \(2\). For example, \(\sqrt{64} = 8\).
import math

result = math.sqrt(25)
print(result)
```
