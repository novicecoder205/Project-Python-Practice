Exercise 3: Celsius to Fahrenheit

Complete the "Celsius to Fahrenheit" exercise.

Instructions:

Write Python code to solve the "Celsius to Fahrenheit" challenge. 

Follow the instructions and make sure your output matches the expected resul

```python
celcius = int(input("What is your celcius score: "))
fahrenheit = (celcius * (9/5)) + 32
print(f"The fahrenheit score of a celsius score of {celcius} is {fahrenheit}")
```

Exercise 4: BMI Calculator

Complete the "BMI Calculator" exercise.

Instructions:

Write Python code to solve the "BMI Calculator" challenge. 

Follow the instructions and make sure your output matches the expected result.

```python
weight = int(input("What is your weight: "))
height = float(input("What is your height: "))
BMI = weight / (height **2)
print(f"your BMI score is: {BMI}")
```
